# Importing essential libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Reading xlsx file and converting into dataframe

```python
df = pd.read_excel('Superstore_USA.xlsx')
df.head(3)
```

# Understanding Data

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

# Missing data analysis

```python
df.isnull().sum()
```

```python
df['Product Base Margin'].fillna(df['Product Base Margin'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

# Univariate Data Analysis

## Order Priority

```python
df['Order Priority'].value_counts()
```

```python
df['Order Priority'].unique()
```

```python
df['Order Priority']=df['Order Priority'].replace('Critical ','Critical')
```

```python
df['Order Priority'].unique()
```

```python
plt.figure(figsize=(10,3))
plt.title("Order Priority analysis",fontsize=24)
plt.xlabel("Order Priority",fontsize=14)
plt.ylabel("Counts",fontsize=14)
sns.countplot(data=df,x='Order Priority')
plt.show()
```

## Ship Mode

```python
df['Ship Mode'].value_counts()
```

```python
x = df['Ship Mode'].value_counts().index
y=df['Ship Mode'].value_counts().values
```

```python
plt.title("Pie chart for shipping mode")
plt.pie(y,labels=x,autopct="%0.2f%%")
plt.show()
```

```python
plt.figure(figsize=(7,3))
sns.countplot(data=df,x='Customer Segment')
plt.show()
```

# Bivariate Data Analysis

```python
plt.figure(figsize=(5,3))
sns.countplot(data=df,x='Ship Mode',hue='Product Category')
plt.show()
```
